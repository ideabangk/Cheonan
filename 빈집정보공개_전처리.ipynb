```python
import pandas as pd
df = pd.read_csv('./00.Data/천안동남구_빈집정보공개.csv')
df
```

```python
df1 = pd.read_csv('./00.Data/천안서북구_빈집정보공개.csv')
df1
```

```python
mdf = pd.concat([df,df1],axis=0)
mdf = mdf.reset_index(drop=True)
mdf['법정동명'] = mdf['소재지'].apply(lambda x: x.split()[2])
mdf.to_csv('./00.Data/천안_빈집정보공개.csv',index=False)
```
