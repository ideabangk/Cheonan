```python
import pandas as pd
df = pd.read_excel("./00.Data/빈집실태조사 결과 현황 조회.xlsx")
df = df.loc[:,:'기타']
df
```

```python
# 등급 전처리
df_grade = df[['지역', '전체 빈집수', '1등급(양호)','2등급(일반)','3등급(불량)','4등급(철거대상)']].loc[1:,:]
df_grade = df_grade.drop([1,23],axis=0).reset_index(drop=True)
df_grade[['전체 빈집수','1등급(양호)', '2등급(일반)']] = df_grade[['전체 빈집수','1등급(양호)', '2등급(일반)']].astype('int64') # 비율계산 위한 부분
df_melted = df_grade.melt(id_vars=["지역", "전체 빈집수"], var_name="등급", value_name="빈집수")
#df_melted.to_csv('./00.Data/동별_빈집등급.csv',index=False)
```

```python
# 용도지역 전처리
df_use = df[['지역','전체 빈집수','주거지역', '상업지역', '공업지역', '녹지지역', '관리지역', '농림지역', '자연환경보전지역']]
df_use = df_use.drop([0,1,23],axis=0).reset_index(drop=True)
df_melt = df_use.melt(id_vars=["지역", "전체 빈집수"], var_name="용도지역", value_name="빈집수")
#df_melt.to_csv('./00.Data/동별_용도지역.csv',index=False)
```

```python
# 주택용도 전처리
df_house = df[['지역','전체 빈집수','소계', '소계.1', '그외주택']]
df_house = df_house.drop([0,1,23],axis=0).reset_index(drop=True)
df_house = df_house.rename(columns={'소계':'단독주택','소계.1':'공동주택','그외주택':'기타주택'})
melted_df = df_house.melt(id_vars=["지역", "전체 빈집수"], var_name="주택용도구분", value_name="빈집수")
melted_df.to_csv('./00.Data/동별_주택용도.csv',index=False)
```

```python
merged = pd.concat([df_grade, df_use, df_house], axis=1)
merged = merged.loc[:, ~merged.T.duplicated()]
merged = merged.loc[:, ~merged.columns.duplicated()]
merged[['관리지역','단독주택','기타주택']] = merged[['관리지역','단독주택','기타주택']].astype('int64')
```

```python
merged.to_csv('./00.Data/군집분석용_데이터.csv',index=False)
```
