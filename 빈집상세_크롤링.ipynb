# 동남구 크롤링

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
import time

# WebDriver 설정
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# 웹페이지 열기
url = "https://binzib.reb.or.kr/binzib/binzibplatform/statisticInfo/binzibPublicInfoStatus.do"
driver.get(url)

# 수동으로 권역 클릭할 시간 대기
print("지도를 클릭하고 엔터를 누르세요.")
input("수동으로 권역을 클릭한 후, 엔터를 눌러주세요...")

all_data = []

while True:
    # 데이터 크롤링
    data_rows = driver.find_elements(By.CSS_SELECTOR, "ul.rightArea")  # <ul> 태그 선택
    
    for row in data_rows:
        # 각 <li> 태그에서 텍스트 추출
        address = row.find_element(By.CSS_SELECTOR, "span.address").text
        apt_type = row.find_element(By.CSS_SELECTOR, "span.aptType").text
        grade = row.find_element(By.CSS_SELECTOR, "span.result").text
        zone = row.find_element(By.CSS_SELECTOR, "span.giyuk").text
        dept = row.find_element(By.CSS_SELECTOR, "span.dept").text
        
        # 데이터 리스트에 추가
        all_data.append([address, apt_type, grade, zone, dept])
    
    # 다음 페이지로 이동하기
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, "li.paginationjs-next")  # "다음" 버튼 찾기
        if "disabled" in next_button.get_attribute("class"):
            break  # "다음" 버튼이 비활성화(마지막 페이지)되면 루프 종료
        else:
            next_button.click()  # "다음" 버튼 클릭
            time.sleep(2)  # 페이지 로드 대기
    except Exception as e:
        print("더 이상 페이지가 없습니다.", e)
        break

# Pandas DataFrame으로 변환 및 CSV 저장
df = pd.DataFrame(all_data, columns=['소재지', '주택유형', '등급', '용도지역', '담당부서'])
df.to_csv("동남구_빈집_크롤링_데이터.csv", index=False)

# 브라우저 닫기
driver.quit()
```
